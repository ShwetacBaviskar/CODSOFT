```python
import tkinter as tk
from datetime import datetime

def respond_to_input(user_input):
    user_input = user_input.strip()

    if user_input.lower() in ["hi", "hello", "hey"]:
        return "Hello! How can I assist you?"

    elif "your name" in user_input.lower():
        return "I'm ChatBot, your virtual assistant."

    elif "good morning" in user_input.lower():
        return "Good Morning. Have a nice day!"

    elif "good afternoon" in user_input.lower():
        return "Good Afternoon. Hope you're having a great day!"

    elif "good evening" in user_input.lower():
        return "Good Evening. How can I assist you?"

    elif "good night" in user_input.lower():
        return "Good Night. Sleep well and take care!"

    elif "how are you" in user_input.lower():
        return "I'm just a program, but I'm functioning perfectly. How about you?"

    elif "time" in user_input.lower():
        current_time = datetime.now().strftime("%H:%M:%S")
        return f"The current time is {current_time}."

    elif "date" in user_input.lower():
        current_date = datetime.now().strftime("%Y-%m-%d")
        return f"Today's date is {current_date}."

    elif "bye" in user_input.lower() or "exit" in user_input.lower():
        return "Goodbye! Have a great day."

    elif "weather" in user_input.lower():
        return "I'm not equipped to check live weather updates right now, but it's always a good idea to carry an umbrella just in case!"

    elif "help" in user_input.lower():
        return "I can answer basic questions about time, date, or engage in casual conversation. Let me know how I can assist you!"

    else:
        return "I'm sorry, I didn't understand that. Could you please rephrase?"

def send_message():
    user_input = user_entry.get()
    if user_input.lower() in ["bye", "exit"]:
        chat_window.insert(tk.END, "You: " + user_input + "\n")
        chat_window.insert(tk.END, "ChatBot: Goodbye! Have a great day.\n")
        root.after(1000, root.destroy)
        return

    chat_window.insert(tk.END, "You: " + user_input + "\n")
    response = respond_to_input(user_input)
    chat_window.insert(tk.END, "ChatBot: " + response + "\n")
    user_entry.delete(0, tk.END)

# Create the GUI application
root = tk.Tk()
root.title("ChatBot")

# Chat window
chat_window = tk.Text(root, height=20, width=50, state=tk.NORMAL)
chat_window.pack(pady=10)

# User input field
user_entry = tk.Entry(root, width=40)
user_entry.pack(pady=5)

# Send button
send_button = tk.Button(root, text="Send", command=send_message)
send_button.pack(pady=5)

# Start the GUI loop
root.mainloop()
```
